```python

# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)

# calculate covariance matrix
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```

```python
#import Auto MPG dataset
df = sns.load_dataset('mpg')
```

```python
df.head()
```

```python
df['company'] = df['name'].apply(lambda x: x.split()[0])
df['model'] = df['name'].apply(lambda x: ' '.join(x.split()[1:]))
df.drop('name', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Select only numeric columns for the correlation matrix
numeric_cols = df.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_cols.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Auto MPG Listings')
plt.xticks(rotation=45, ha='right')  # Rotate x labels
plt.yticks(rotation=0)  # Keep y labels horizontal
plt.show()
```

The heatmap allows us to check if there is any correlation between the variables. From the above chart, there is a strong correlation between displacement and weight,cylinders and horsepower. There is a slight correlation between model year and mpg and acceleration.

```python
#Create a correlation chart to show degree of relationship between Acceleration and MPG
plt.figure(figsize=(12, 10))
plt.scatter(df['mpg'], df['acceleration'], color='blue')
plt.xlabel('MPG')
plt.ylabel('Acceleration')
plt.title('Correlation Chart Acceleration and MPG')
plt.show()
```

As the acceleration increases the MPG also increases denoting a positive correlation.

```python
# Relationship between displacement and weight
plt.figure(figsize=(12, 10))
plt.scatter(df['displacement'], df['weight'], color='green')
plt.xlabel('displacement')
plt.ylabel('weight')
plt.title('Correlation Chart Displacement and Weight')
plt.show()
```

```python
#Show relationship between horsepower and cylinders
plt.figure(figsize=(12, 10))
plt.scatter(df['horsepower'], df['cylinders'], color='brown')
plt.xlabel('horsepower')
plt.ylabel('cylinders')
plt.title('Correlation CHart Horsepower and Cylinders')
plt.show()
```
