In the next section, analysis of mpg of USA made cars and their model_years.

```python
#group all the USA made cars
model_usa =df2[df2['origin'] == 'usa']
```

```python
model_usa.head()
```

```python
model_usa.groupby('model_year').describe()['mpg']
```

In the above summary, we are presented with mpg statistical summaries of cars that originated in the USA between 1970 - 1982. For instance, there are 22 cars in the dataset from 1970, with an average mpg of 15, a minimum of 9 mpg and a maximum of 22.

```python
plt.figure(figsize=(10,6))
plt.axhline(model_usa.mpg.mean(), color='r', linestyle='--', linewidth=2)
sns.boxplot(x='model_year', y='mpg', data=model_usa, palette='pastel')
plt.show()
```

The dotted line represents the mean mpg for USA made cars over the last decade. From 1978, the mean mpg for each year has increased exponentially from the standard mean of 20 mpg represented by the dotted line. It is importatnt to note that in 78, there were significant outliers in mpg recorded that year.

```python
plt.figure(figsize=(10,6))
df2.groupby('weight_level').describe()['mpg'].plot(kind='bar')
plt.show()
```
