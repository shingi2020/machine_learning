```python
total_count = len(dataset)
print(total_count)
```

```python
train_size = int(total_count * 0.7)
val_size = int(total_count * 0.15)
test_size = int(total_count * 0.15)
```

```python
#complete_dataset = dataset.shuffle(total_count, seed=42)

train_dataset = dataset.take(train_size)
remaining_dataset = dataset.skip(train_size)
val_dataset = remaining_dataset.take(val_size)
test_dataset = remaining_dataset.skip(val_size)
```

```python

train_dataset = dataset['train'].take(train_size)
remaining_dataset = dataset['train'].skip(train_size)
val_dataset = remaining_dataset.take(val_size)
test_dataset = remaining_dataset.skip(val_size)
```

```python
#Normalize the dataset
train_data = np.array(list(train_dataset.as_numpy_iterator()))


mean = tf.reduce_mean(train_data)
std = tf.math.reduce_std(train_data)

def normalize(x):
    return (x - mean) / std

normalized_train_dataset = train_dataset.map(normalize)
normalized_val_dataset = val_dataset.map(normalize)
normalized_test_dataset = test_dataset.map(normalize)

```

```python
def cast_to_float32(x):
    # Iterate over the dictionary items and cast tensors to float32
    for key, value in x.items():
        if isinstance(value, tf.Tensor) and value.dtype != tf.float32:
            x[key] = tf.cast(value, tf.float32)
    return x

train_dataset = train_dataset.map(cast_to_float32)
val_dataset = val_dataset.map(cast_to_float32)
test_dataset = test_dataset.map(cast_to_float32)

def divide_by_255(x):
    # Iterate over the dictionary items and divide tensors by 255.0
    for key, value in x.items():
        if isinstance(value, tf.Tensor):
            x[key] = value / 255.0
    return x

train_dataset = train_dataset.map(divide_by_255)
val_dataset = val_dataset.map(divide_by_255)
test_dataset = test_dataset.map(divide_by_255)
```
